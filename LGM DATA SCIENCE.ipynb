```python
Customer_id=Cstmr_details["Customer_ID"]
store_Code=Store_deatils["Store_Code"][:31]
```

```python
All_Customer_store=[]
for cst in Customer_id:
    for st in store_Code:
        All_Customer_store.append(str(cst)+"#"+str(st))
```

```python
Cstmr_trstn["Customer_store"]=Cstmr_trstn["Customer_ID"].apply(lambda x:str(x))+"#"+Cstmr_trstn["Store_Code"].apply(lambda x:str(x))
```

```python
fav_customer_store=[i for i in Cstmr_trstn["Customer_store"].value_counts().index]
```

```python
Customer_store_dict=dict(list(zip(All_Customer_store,np.zeros(len(All_Customer_store)))))
```

```python
fav_dict=dict(list(zip(fav_customer_store,np.zeros(len(fav_customer_store)))))
```

```python
for val in Customer_store_dict:
    if val in fav_dict:
        Customer_store_dict[val]=1
        
```

```python
a,b=zip(*list(Customer_store_dict.items()))
```

```python
Train=pd.DataFrame()
Train["Customer_store"]=a
Train["Target"]=b
```

```python
Train.head()
```

```python
Train["Customer_ID"]=Train["Customer_store"].apply(lambda x:int(x.split("#")[0]))
Train["Store_Code"]=Train["Customer_store"].apply(lambda x:int(x.split("#")[1]))
```

```python
X_train["Region_Code"]=X_train["Region_Code"].astype("object")
X_test["Region_Code"]=X_test["Region_Code"].astype("object")
```

```python
## label encoding
for val in cat_var:
    print(val)
    final_col=X_train[val].append(X_test[val])
    final_col=le.fit_transform(final_col)
    X_train[val]=final_col[:X_train.shape[0]]
    X_test[val]=final_col[X_train.shape[0]:]
  
```

```python
X_tr.reset_index(inplace=True,drop=True)
y_tr.reset_index(inplace=True,drop=True)
X_val.reset_index(inplace=True,drop=True)
y_val.reset_index(inplace=True,drop=True)
```
